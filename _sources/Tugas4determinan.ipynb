#  Determinan






Determinan adalah suatu bilangan real yang diperoleh dari suatu proses dengan aturan tertentu terhadap matriks bujur sangkar
. Secara lebih umum, determinan merupakan nilai skalar yang hanya memiliki besaran atau nilai, yang dihitung dari unsur-unsur suatu matriks persegi
. Dalam matematika, determinan adalah sebuah fungsi tertentu yang menghubungkan suatu bilangan real dengan suatu matriks bujur sangkar
.

Determinan matriks A ditulis dengan tanda det(A), det A, atau |A|, dan dapat dianggap sebagai faktor penskalaan transformasi yang digambarkan oleh matriks
. Nilai determinan ini dapat menentukan beberapa sifat dari matriks dan pemetaan linear yang direpresentasikan oleh matriks tersebut.

Dalam konteks matriks, determinan juga dapat dihitung dengan menggunakan metode ekspansi kofaktor baris pertama
. Jika nilai determinan suatu matriks bujur sangkar adalah 0, maka matriks tersebut disebut singular, yang berarti matriks tersebut tidak memiliki invers. Sebaliknya, jika nilai determinan matriks bujur sangkar tidak sama dengan 0, maka matriks tersebut disebut nonsingular, yang artinya matriks tersebut memiliki invers
.

#  Syarat mencari determinan matriks

Determinan dari suatu matriks dapat dihitung dengan beberapa syarat yang perlu dipenuhi. Berikut adalah beberapa syarat umum untuk mencari determinan matriks:

Matriks Persegi: Determinan hanya dapat dihitung untuk matriks yang merupakan matriks persegi, yaitu matriks yang memiliki jumlah baris sama dengan jumlah kolomnya
.
Ukuran Matriks: Semakin besar ukuran matriks, semakin kompleks cara mencari determinannya. Matriks ordo 2x2 merupakan matriks paling sederhana, sehingga cara mencari determinannya juga sederhana
.
Metode Perhitungan: Terdapat beberapa metode untuk menghitung determinan matriks, seperti metode pertama yang dihitung dengan rumus tertentu, dan metode kedua yang mereduksi determinan menjadi ordo 2
.
Matriks Invers: Determinan matriks juga berkaitan dengan matriks invers. Misalnya, jika suatu matriks bujur sangkar A dikalikan terhadap inversnya yaitu matriks bujur sangkar A1, maka menghasilkan matriks identitas pada operasi perkalian matriks
.

Dengan memenuhi syarat-syarat tersebut, determinan matriks dapat dihitung sesuai dengan rumus dan metode yang berlaku untuk matriks tersebut.

#  Sifat sifat determinan

Determinan merupakan suatu nilai yang terkait dengan matriks dan memiliki beberapa sifat yang penting dalam matematika. Beberapa sifat-sifat determinan yang perlu diketahui antara lain:

1. Sifat Determinan Matriks Identitas:

Identitas merupakan matriks diagonal, sehingga determinannya sama dengan hasil kali dari semua entri diagonalnya, yang semuanya bernilai 1. Sehingga, determinan dari matriks identitas adalah 1.

2. Sifat Determinan Matriks Transpose:

Determinan dari matriks transpose sama dengan determinan dari matriks aslinya, yaitu det(A) = det(A^T).

3. Sifat Determinan Matriks Skalar:

Jika A adalah matriks berukuran n x n, maka determinan dari kA adalah kn det(A), di mana k adalah skalar
.

4. Sifat Determinan Matriks Segitiga:

Jika A adalah matriks segitiga atas, segitiga bawah, atau diagonal, maka determinannya adalah hasil kali dari elemen-elemen diagonal utamanya, yaitu det(A) = a11a22a33...ann
.

5. Sifat Determinan Matriks Invers:

Determinan dari invers matriks A dapat dihitung dengan rumus det(A^-1) = 1/det(A).

6. Sifat Determinan Matriks Penjumlahan:

Determinan dari penjumlahan dua matriks A dan B sama dengan determinan dari matriks A ditambah determinan dari matriks B, yaitu det(A+B) = det(A) + det(B)
.

7. Sifat Determinan Matriks Perkalian:

Determinan dari hasil perkalian dua matriks A dan B sama dengan perkalian determinan matriks A dan B, yaitu det(AB) = det(A) * det(B)
.

#   Menghitung Kofaktor, Determinan dan Invers suatu matriks

```python
import numpy as np

# Menghitung kofaktor suatu elemen dalam matriks
def cofactor(matrix, i, j):
    return (-1) ** (i + j) * np.linalg.det(np.delete(np.delete(matrix, i, axis=0), j, axis=1))

# Menghitung matriks kofaktor dari matriks input
def cofactor_matrix(matrix):
    return np.array([[cofactor(matrix, i, j) for j in range(matrix.shape[1])] for i in range(matrix.shape[0])])

# Menghitung determinan matriks
def determinant(matrix):
    return np.linalg.det(matrix)

# Contoh matriks
A = np.array([[3,1,3],
              [2,1,2],
              [4,2,1]])

# Menghitung matriks kofaktor
C = cofactor_matrix(A)
print("Matriks Kofaktor:")
print(C)

# Menghitung determinan
det_A = determinant(A)
print("\nDeterminan:", det_A.round())

# Menghitung invers matriks
def inverse(matrix):
    det = determinant(matrix)
    if det == 0:
        raise ValueError("Determinan nol, matriks tidak dapat diinvers")
    else:
        return cofactor_matrix(matrix).T / det

# Menghitung invers dari matriks A
A_inv = inverse(A)
print("\nInvers Matriks A:")
print(A_inv.round(2))


```

#  Determinan dengan Ekspansi Kolom Kofaktor

```python
import numpy as np

def minor(matriks, i, j):
    """Menghitung minor dari elemen (i, j) dalam matriks."""
    return np.linalg.det(np.delete(np.delete(matriks, i, axis=0), j, axis=1))

def kofaktor(matriks, i, j):
    """Menghitung kofaktor dari elemen (i, j) dalam matriks."""
    return ((-1) ** (i + j)) * minor(matriks, i, j)

def determinan_dengan_kolom(matriks, kolom):
    """Menghitung determinan matriks menggunakan ekspansi kofaktor pada kolom tertentu."""
    n = len(matriks)
    return sum(matriks[i][kolom] * kofaktor(matriks, i, kolom) for i in range(n))

# Contoh penggunaan:
matriks = np.array([[3, 1, 3],
                    [2, 1, 2],
                    [4, 2, 1]])
det_matriks = determinan_dengan_kolom(matriks, 1)  # Menggunakan kolom kedua (indeks 1)
print(f"Determinan matriks dengan menggunakan kolom kedua adalah: {det_matriks.round()}")
```

#  Refleksi Terhadap Sumbu X dan Y

```python
import numpy as np

def refleksi_terhadap_sumbu_x(koordinat):
    # Matriks refleksi terhadap sumbu x
    Fx = np.array([[1, 0],
                   [0, -1]])

    # Melakukan refleksi terhadap sumbu x
    koordinat_setelah = np.dot(Fx, koordinat)
    return koordinat_setelah

def refleksi_terhadap_sumbu_y(koordinat):
    # Matriks refleksi terhadap sumbu y
    Fy = np.array([[-1, 0],
                   [0, 1]])

    # Melakukan refleksi terhadap sumbu y
    koordinat_setelah = np.dot(Fy, koordinat)
    return koordinat_setelah

# Contoh koordinat objek
koordinat_objek = np.array([[2],
                             [3]])

# Refleksi terhadap sumbu x
hasil_refleksi_x = refleksi_terhadap_sumbu_x(koordinat_objek)

# Refleksi terhadap sumbu y
hasil_refleksi_y = refleksi_terhadap_sumbu_y(koordinat_objek)

print("Koordinat objek sebelum refleksi:")
print(koordinat_objek)

print("\nKoordinat objek setelah refleksi terhadap sumbu x:")
print(hasil_refleksi_x)

print("\nKoordinat objek setelah refleksi terhadap sumbu y:")
print(hasil_refleksi_y)
```

#  2. Shearing

Shearing adalah transformasi yang mengubah bentuk objek tanpa mengubah luasnya. Dalam ruang dua dimensi, shearing dapat terjadi sepanjang sumbu x atau sumbu y. Transformasi shearing menggeser satu koordinat secara proporsional terhadap yang lain, menciptakan efek miring atau deformasi pada bentuk asli.
Transformasi shearing adalah jenis transformasi geometri di mana titik-titik suatu objek digeser sepanjang sumbu tertentu, tanpa mengubah ukuran objek tersebut.

```python
def shearing_transform(points, shear_factor):
    """
    Melakukan transformasi shearing pada sebuah objek berdasarkan faktor shearing yang diberikan.

    Args:
    - points: List dari tuple koordinat titik-titik objek.
    - shear_factor: Tuple berisi faktor shearing untuk sumbu x dan sumbu y.

    Returns:
    - List dari tuple koordinat titik-titik objek setelah transformasi shearing.
    """
    return [(x + shear_factor[0] * y, y + shear_factor[1] * x) for x, y in points]

# Contoh penggunaan
if __name__ == "__main__":
    # Definisikan titik-titik objek awal (dalam format list dari tuple)
    points = [(0, 0), (1, 0), (1, 1), (0, 1)]

    # Faktor shearing (misalnya, shearing pada sumbu x sebesar 0.5 dan pada sumbu y sebesar 0.3)
    shear_factor = (0.5, 0.3)

    # Melakukan transformasi shearing
    transformed_points = shearing_transform(points, shear_factor)

    # Menampilkan hasil transformasi
    print("Titik-titik setelah transformasi shearing:")
    for point in transformed_points:
        print(point)
```

Rotasi mempertahankan jarak antara titik-titik dan asal, serta sudut antara garis-garis.Transformasi rotasi adalah jenis transformasi geometri di mana suatu objek diputar sekitar titik tertentu yang disebut pusat rotasi. Dalam rotasi, setiap titik objek diputar sekitar pusat rotasi sebesar sudut tertentu dalam arah tertentu. kode Python untuk melakukan rotasi dua dimensi terhadap suatu titik (x, y) sebesar sudut tertentu (dalam radian) searah jarum jam:

```python
import math

def rotate_point(x, y, angle):
    # Mengubah sudut rotasi menjadi radian
    angle_rad = math.radians(angle)

    # Menghitung koordinat baru setelah rotasi
    x_rotated = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    y_rotated = x * math.sin(angle_rad) + y * math.cos(angle_rad)

    return x_rotated, y_rotated

# Contoh penggunaan
x = 1
y = 1
angle = 45  # Sudut rotasi dalam derajat

x_rotated, y_rotated = rotate_point(x, y, angle)

print("Koordinat setelah rotasi:", x_rotated, y_rotated)
```

#  Rotasi 90 Derajat

```python
import numpy as np

# Membuat matriks 2x2 dengan nilai awal
matriks = np.array([[1, 0], [0, 1]])

# Mendefinisikan matriks rotasi untuk rotasi 90 derajat searah jarum jam
theta = np.radians(90)  # Mengubah derajat menjadi radian
rotasi_matriks = np.array([[np.cos(theta), -np.sin(theta)],
                           [np.sin(theta), np.cos(theta)]])

# Melakukan rotasi matriks
matriks_rotasi = np.dot(rotasi_matriks, matriks)

# Menampilkan matriks setelah rotasi
print("Matriks setelah rotasi 90 derajat searah jarum jam:")
print(matriks_rotasi.round())
```

#  4. Squeezing

Squeezing adalah transformasi di mana skala dalam satu arah berbeda dari skala dalam arah lainnya. Contohnya, sebuah objek dapat diperpanjang dalam arah x dan diperkecil dalam arah y, atau sebaliknya. Matriks yang digunakan untuk squeezing adalah matriks skala dengan faktor skala berbeda untuk setiap arah.  Transformasi squeezing adalah jenis transformasi geometri di mana objek dipersempit atau diperlebar dalam satu atau lebih arah. Hal ini dilakukan dengan memperbesar atau memperkecil jarak antara titik-titik objek sepanjang sumbu tertentu.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan squeezing horizontal
def squeeze_horizontal(points, factor):
    return points * np.array([factor, 1])

# Fungsi untuk melakukan squeezing vertikal
def squeeze_vertical(points, factor):
    return points * np.array([1, factor])

# Titik-titik objek awal
points = np.array([[1, 1], [2, 2], [3, 1]])

# Faktor squeezing horizontal dan vertikal
horizontal_factor = 0.5
vertical_factor = 1.5

# Lakukan transformasi squeezing horizontal dan vertikal
squeezed_horizontal_points = squeeze_horizontal(points, horizontal_factor)
squeezed_vertical_points = squeeze_vertical(points, vertical_factor)

# Plot hasil transformasi
plt.scatter(points[:, 0], points[:, 1], label='Original Points')
plt.scatter(squeezed_horizontal_points[:, 0], squeezed_horizontal_points[:, 1], label='Squeezed Horizontal Points')
plt.scatter(squeezed_vertical_points[:, 0], squeezed_vertical_points[:, 1], label='Squeezed Vertical Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Squeezing Transformation')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
